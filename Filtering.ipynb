```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('image1.jpg', -1)
plt.imshow(image),plt.title('Original')
plt.show()
```

```python
kernel = np.ones((5,5),np.float32)/25
output1 = cv2.filter2D(image,-1,kernel)
cv2.startWindowThread()
cv2.imwrite("OutputImages/Averaging.jpg",output1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blur = cv2.blur(image,(5,5))
cv2.startWindowThread()
cv2.imshow("OutputImages/Blur.jpg",blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gaussianBlur = cv2.GaussianBlur(image,(5,5),0)
cv2.startWindowThread()
cv2.imshow("OutputImages/Gaussian Blur",gaussianBlur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
bilateral = cv2.bilateralFilter(image,9,75,75)
cv2.startWindowThread()
cv2.imshow("OutputImages/Median Blur",bilateral)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
